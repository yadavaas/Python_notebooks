# Pie Chart

[link for Pie chart](https://mldoodles.com/matplotlib-pie-chart/)

```python
import matplotlib.pyplot as plt
```

```python
class1_students = [45, 15, 35, 25, 30]
plt.pie(class1_students);
# plt.show()
```

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]

plt.pie(class1_students, labels = classes);
# plt.show()
```

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
explode = [0,0.5,0,0.9,0]
colors = ["c", 'b','r','y','g']
textprops = {"fontsize":12,"color":"k",
"weight":"bold"}

plt.pie(class1_students, 
        labels = classes, 
        explode = explode, 
        colors =colors, 
        autopct = "%0.0f%%", 
        shadow = True, 
        radius = 1,
       startangle = 90, 
        textprops =textprops)
plt.show()
```

```python
plt.figure(figsize = (5,5))
wedgeprops = {"linewidth": 5,'width':.7, "edgecolor":"m"}
plt.pie(
        class1_students, 
        labels = classes, 

        colors = colors, 
        autopct = "%0.1f%%", 
        pctdistance = 0.8,
        shadow =True, 
        labeldistance = 1.1,
        startangle = 150,
        radius = 1.5, 
        counterclock = False,
        wedgeprops = wedgeprops,
        textprops = textprops,
        
        frame=0,
        rotatelabels=True
        ) 
plt.show()
```

```python
plt.figure(figsize = (5,5))
explode = [0.03,0,0.1,0.3,0.5]
colors = ["c", 'b','r','y','g']
textprops = {"fontsize":20}

plt.pie(class1_students, labels = classes, explode = explode, 
        colors =colors, autopct = "%0.3f%%", shadow = True, radius = 1.9,
       startangle = 270, textprops =textprops)
# plt.legend(loc=2)
plt.legend(bbox_to_anchor=(2, 1), loc='upper right', borderaxespad=0)
plt.show()
```
